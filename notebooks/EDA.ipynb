```python
PATH_EIVP_DATASET_MESURES = "../data/EIVP_DATASET_MESURES.xlsx"
# PATH_DFTRIE = "../data/dftrie.xlsx"
PATH_METEO = "../data/donnees-synop-essentielles-omm.xlsx"
```

```python
import seaborn as sns
import pandas as pd

data= pd.read_excel(PATH_EIVP_DATASET_MESURES)
#data.describe()
data
```

```python
data['sent_at'] = pd.to_datetime(data['sent_at'])
data = data.set_index('sent_at')
data.index
```

```python
from datetime import datetime
data[data.index < datetime.strptime('2021-04-13 20:00:00', "%Y-%m-%d %H:%M:%S")]
```

```python
# data = data.reset_index()
# data = data.set_index('sent_at')
df = data.resample('3H', origin=datetime.strptime('2021-04-13 17:00:00', "%Y-%m-%d %H:%M:%S")).mean()
# data = data.reset_index()
```

```python
df
```

```python
meteo = pd.read_excel(PATH_METEO)
meteo.head()
```

```python
meteo["Date"] = pd.to_datetime(meteo["Date"])
meteo['sent_at'] = pd.to_datetime(meteo['Date'].dt.tz_localize(None))
```

```python
meteo.head()
```

```python
df_merge = df.merge(meteo, on='sent_at')
df_merge.head()
```
