```python
import seaborn as sns
import pandas as pd

data= pd.read_excel(PATH_EIVP_DATASET_MESURES)
#data.describe()
data
```

```python
data['sent_at'] = pd.to_datetime(data['sent_at'])
data = data.set_index('sent_at')
data.index
```

```python
from datetime import datetime
data[data.index < datetime.strptime('2021-04-13 20:00:00', "%Y-%m-%d %H:%M:%S")]
```

```python
# data = data.reset_index()
# data = data.set_index('sent_at')
df = data.resample('3H', origin=datetime.strptime('2021-04-13 17:00:00', "%Y-%m-%d %H:%M:%S")).mean()
# data = data.reset_index()
```

```python
df
```

```python
meteo = pd.read_excel(PATH_METEO)
meteo.head()
```

```python
meteo["Date"] = pd.to_datetime(meteo["Date"])
meteo['sent_at'] = pd.to_datetime(meteo['Date'].dt.tz_localize(None))
```

```python
meteo.head()
```

```python
df_merge = df.merge(meteo, on='sent_at')
df_merge.head()
```

```python
dflight = df_merge[['sent_at','temp','TempÃ©rature (Â°C)']]
dflight.rename(columns = {'TempÃ©rature (Â°C)': 'temp_ext'}, inplace=True)
dflight
```

```python
import seaborn as sns
sns.relplot(x="temp", y="temp_ext", data=dflight)
```

```python
sns.lmplot(x="temp", y="temp_ext", data=dflight)
```

```python
dftrie = dflight.sort_values(by = 'temp_ext')
dftrie
```

```python
sns.lmplot(x="temp_ext", y="temp", data=dftrie)
```

```python
from numpy import *

dftrie['exp(temp)']= dftrie['temp'].apply(exp)
dftrie
```

```python
sns.lmplot(x="temp_ext", y="exp(temp)", data=dftrie)
```
